```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
import json
```

## Step 1 — Create a unified dataframe

In this step, all CSV files from different countries were concatenated into a single dataframe.  
A new column named **`country`** was added to identify the country each file belongs to.  
This step ensures that all the data is combined in a structured and consistent format for further analysis.

```python
#1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

# Path to dataset folder: go one level up (..) then into 'trendingYT'
data_path = os.path.join("..", "trendingYT")

# Find all CSV files that end with "videos.csv"
csv_files = glob.glob(os.path.join(data_path, "*videos.csv"))
print("Found CSV files:")
for f in csv_files:
    print(" -", f)

dfs = []

for file in csv_files:
    # Extract country code from the filename (first two letters, e.g., US, CA, DE)
    country_code = os.path.basename(file)[:2].upper()
    print(f"\nReading {file} -> country = {country_code}")
    
    # Read CSV (latin-1 encoding avoids common decoding errors)
    df = pd.read_csv(file, encoding="latin-1")
    
    # Add a new column indicating the country of the file
    df["country"] = country_code
    
    # Store the dataframe for later concatenation
    dfs.append(df)

# Concatenate all dataframes into a single dataframe
all_videos = pd.concat(dfs, ignore_index=True)

print("\nFinal shape:", all_videos.shape)

# Display the first few rows of the final dataframe
all_videos.head()
```

## Dataset Overview
A quick look at the dataset structure before applying transformations.

```python
all_videos.head()
all_videos.info()
all_videos.describe()
```

## Step 2 — Extract videos with no tags

The goal of this step was to identify all videos that contain *no meaningful tag information*.  
Since YouTube's dataset does not store missing tags as `NaN`, we handled cases such as:

- empty strings `""`
- whitespace-only entries
- the placeholder `"[none]"`

All such entries were collected into a separate dataframe for further inspection.

```python
# Part 2 - Extract all videos that have no tag

# Some videos have no tags: tags column can be NaN, empty string, or "[none]"
no_tag_mask = (
    all_videos["tags"].isna() |                          # missing value
    all_videos["tags"].str.strip().eq("") |              # empty string
    all_videos["tags"].str.strip().eq("[none]")          # explicit [none] marker
)

# Create a new dataframe with only videos that have no tag
videos_no_tags = all_videos[no_tag_mask].copy()

print("Number of videos with no tags:", videos_no_tags.shape[0])
videos_no_tags.head()
```

## Step 3 — Total number of views per channel

Here, the total number of views for each YouTube channel was computed.  
The dataframe was grouped by **`channel_title`**, and the `views` column was aggregated using the sum.  
This allows us to compare the relative popularity of channels across the dataset.

```python
# Part 3 - Total number of views for each channel

# Group by channel_title and sum all views
channel_total_views = (
    all_videos.groupby("channel_title")["views"]
    .sum()
    .reset_index()
    .sort_values(by="views", ascending=False)
)

print("Number of channels:", channel_total_views.shape[0])
channel_total_views.head(10)
```

## Step 4 — Excluded dataframe and data cleaning

In this step, all videos that were considered *invalid* or *incomplete* were removed from the main dataframe.  
These include videos where:

- comments are disabled  
- ratings are disabled  
- the video is marked as removed or in error  

All such rows were saved in a new dataframe named **`excluded`**, while the remaining clean data was kept in the main dataframe for further analysis.

```python
# Part 4 - Build 'excluded' dataframe and remove those rows from all_videos

# Make sure boolean-like columns are properly interpreted as booleans
bool_cols = ["comments_disabled", "ratings_disabled", "video_error_or_removed"]

for col in bool_cols:
    # Convert values like "True"/"False"/"true"/"false"/"1"/"0" into real booleans
    all_videos[col] = (
        all_videos[col]
        .astype(str)               # convert to string
        .str.strip()               # remove spaces
        .str.lower()               # normalize case
        .isin(["true", "1", "yes"])  # treat these as True, all others as False
    )

# Build the mask:
# - rows where both comments and ratings are disabled
#   OR
# - rows where video_error_or_removed is True
excluded_mask = (
    (all_videos["comments_disabled"] & all_videos["ratings_disabled"]) |
    (all_videos["video_error_or_removed"])
)

# DataFrame with all excluded rows
excluded = all_videos[excluded_mask].copy()
print("Number of rows in 'excluded':", excluded.shape[0])

# Remove those rows from the main dataframe
all_videos = all_videos[~excluded_mask].copy()
print("Number of rows remaining in 'all_videos':", all_videos.shape[0])

# Quick check
excluded.head()
```

## Step 5 — Add the like_ratio feature

A new feature named **`like_ratio`** was created to quantify engagement.  
It is defined as the ratio between likes and dislikes:

\[
\text{like_ratio} = \frac{\text{likes}}{\text{dislikes}}
\]

This metric provides a normalized way to compare video popularity beyond raw like counts.

```python
# Part 5 - Add like_ratio column

# Avoid division by zero: replace 0 dislikes with NaN
all_videos["dislikes_clean"] = all_videos["dislikes"].replace(0, pd.NA)

# Compute ratio: likes / dislikes
all_videos["like_ratio"] = all_videos["likes"] / all_videos["dislikes_clean"]

# Drop helper column if you don't need it later
all_videos.drop(columns=["dislikes_clean"], inplace=True)

# Preview
all_videos[["likes", "dislikes", "like_ratio"]].head()
```

## Step 6 — Cluster publish times into 10-minute intervals

The `publish_time` column was first converted into a proper datetime format.  
Then, each timestamp was rounded down to the nearest **10-minute interval** using the `floor("10min")` method.  
This prepares the data for time-based aggregation and trend analysis in the following steps.

```python
# Step 6 — Cluster publish time into 10-minute intervals (time-of-day only)

# 1) Convert publish_time to datetime format.
#    If a value cannot be parsed, it will be converted to NaT (Not a Time).
#    utc=True ensures consistent timezone handling.
all_videos["publish_time"] = pd.to_datetime(
    all_videos["publish_time"],
    errors="coerce",
    utc=True
)

# 2) Round down each timestamp to the nearest 10-minute boundary.
#    For example, 02:23 → 02:20.
pt_floor = all_videos["publish_time"].dt.floor("10min")

# 3) Compute the upper bound of each 10-minute interval.
#    For example, if the lower bound is 02:20, the upper bound becomes 02:30.
pt_next = pt_floor + pd.Timedelta(minutes=10)

# 4) Create a readable interval label in the format "HH:MM to HH:MM".
#    Example: "02:20 to 02:30".
all_videos["publish_interval"] = (
    pt_floor.dt.strftime("%H:%M") +
    " to " +
    pt_next.dt.strftime("%H:%M")
)

# 5) Display the first few rows to verify the transformation.
all_videos[["publish_time", "publish_interval"]].head()
```

## Step 7 — Interval statistics (video count, avg likes, avg dislikes)

For each 10-minute publish interval, the following metrics were computed:

- total number of videos  
- average number of likes  
- average number of dislikes  

This gives insight into how upload timing correlates with video performance.

```python
# Part 7 - Statistics per 10-minute publish interval

interval_stats = (
    all_videos.groupby("publish_interval")
    .agg(
        videos_count=("video_id", "count"),
        avg_likes=("likes", "mean"),
        avg_dislikes=("dislikes", "mean")
    )
    .reset_index()
)

interval_stats.head()
```

```python
# Select every 6th interval label.
# Since each interval represents 10 minutes, 6 intervals correspond to 1 hour.
# This reduces overcrowding on the x-axis.
ticks = interval_stats["publish_interval"][::6]

# Create a new figure with a wider size for better readability.
plt.figure(figsize=(14,6))

# Plot the number of videos for each 10-minute publish interval.
plt.plot(
    interval_stats["publish_interval"],
    interval_stats["videos_count"]
)

# Display only selected ticks (one per hour) and rotate them for clarity.
plt.xticks(ticks=ticks, rotation=45)

# Add a descriptive title to the chart.
plt.title("Number of Videos per 10-minute Interval")

# Adjust layout to prevent overlapping of labels and margins.
plt.tight_layout()

# Render the plot.
plt.show()
```

## Step 8 — Count number of videos per tag

The `tags` column contains multiple tags separated by the `|` character.  
The tags were split into lists and then **exploded** so that each tag appears in its own row.  
After cleaning empty values, the number of videos associated with each tag was counted.

```python
# Part 8 - Count number of videos per tag

# Replace [none] with empty lists
all_videos["tags"] = all_videos["tags"].replace("[none]", "")

# Split tags into list
all_videos["tag_list"] = all_videos["tags"].str.split("|")

# Explode into multiple rows
tag_exploded = all_videos.explode("tag_list")

# Remove empty tags
tag_exploded = tag_exploded[tag_exploded["tag_list"].str.strip() != ""]

# Count videos per tag
tag_counts = (
    tag_exploded.groupby("tag_list")["video_id"]
    .count()
    .reset_index(name="video_count")
    .sort_values("video_count", ascending=False)
)

tag_counts.head()
```

## Step 9 — Find the most frequent tags

Using the tag counts computed in Step 8, the tags with the highest number of associated videos were identified.  
This highlights the most common themes or categories across the trending dataset.

```python
# Part 9 - Tags with the largest number of videos

top_tags = tag_counts.head(10)
top_tags
```

```python
most_common_tag = tag_counts.iloc[0]
most_common_tag
```

```python
plt.figure(figsize=(10,6))
plt.barh(top_tags["tag_list"], top_tags["video_count"])
plt.xlabel("Number of Videos")
plt.ylabel("Tag")
plt.title("Top 10 Most Frequent Tags")
plt.gca().invert_yaxis()
plt.show()
```

### Interpretation
The tag **"funny"** appears as the most frequent tag, suggesting that humor-related content is highly represented in the global trending dataset.  
This aligns with the general observation that entertainment-focused content tends to trend more often.

Interestingly, several high-frequency tags belong to music and entertainment categories, indicating a strong presence of pop culture in the trending lists.

This result connects well with Step 7, where intervals with higher upload density also showed strong engagement metrics, suggesting that popular tags coincide with high-activity time periods.

## Step 10 — Average like_ratio for each (tag, country) pair

For each combination of **tag** and **country**, the average `like_ratio` was computed.  
This allows us to compare how specific tags perform in different regions, providing insights into country-specific trends and preferences.

```python
# Part 10 - Average like_ratio for each (tag, country) pair

tag_country_ratio = (
    tag_exploded.groupby(["tag_list", "country"])["like_ratio"]
    .mean()
    .reset_index()
    .sort_values("like_ratio", ascending=False)
)

tag_country_ratio.head()
```

## Step 11 — Top video per (trending_date, country) pair

In this step, for each combination of **`trending_date`** and **`country`**,  
the video with the largest number of views was selected.

To do this, the dataframe was sorted by `trending_date`, `country`, and `views` (in descending order),  
and then, for each group, only the top row (the most viewed video) was retained.

This allows us to see which videos dominated the trending list on a given date in each country.

```python
# Step 11 - For each (trending_date, country) pair, find the video with the largest number of views

# In case trending_date is still a string, we can keep it as is for grouping.
# If it was already converted to datetime earlier, this still works the same.

top_video_by_date_country = (
    all_videos
    .sort_values(["trending_date", "country", "views"], ascending=[True, True, False])
    .groupby(["trending_date", "country"], as_index=False)
    .head(1)
)

top_video_by_date_country.head()
```

## Step 12 — Split trending_date into year, month, and day

Here, the **`trending_date`** column was converted into a proper datetime format.  
Then, three new columns were created:

- `trending_year`  
- `trending_month`  
- `trending_day`  

These additional fields make it easier to aggregate and analyze results over different temporal granularities (e.g. by month or year).

```python
# Step 12 — Fix trending_date format and extract year, month, day

# Convert trending_date using known YouTube format: YY.DD.MM
all_videos["trending_date"] = pd.to_datetime(
    all_videos["trending_date"],
    format="%y.%d.%m",
    errors="coerce"
)

# Extract date components
all_videos["trending_year"] = all_videos["trending_date"].dt.year
all_videos["trending_month"] = all_videos["trending_date"].dt.month
all_videos["trending_day"] = all_videos["trending_date"].dt.day

all_videos[["trending_date", "trending_year", "trending_month", "trending_day"]].head()
```

```python
# Step 13 - For each (month, country) pair, find the video with the largest number of views

top_video_by_month_country = (
    all_videos
    .sort_values(["trending_year", "trending_month", "country", "views"],
                 ascending=[True, True, True, False])
    .groupby(["trending_year", "trending_month", "country"], as_index=False)
    .head(1)
)

top_video_by_month_country.head()
```

```python
monthly_top_counts = (
    top_video_by_month_country
    .groupby("country")["video_id"]
    .count()
    .reset_index(name="months_with_top_video")
    .sort_values("months_with_top_video", ascending=False)
)

plt.figure(figsize=(8,5))
plt.barh(monthly_top_counts["country"], monthly_top_counts["months_with_top_video"])
plt.xlabel("Number of (year, month) combinations with a top video")
plt.ylabel("Country")
plt.title("Count of monthly top videos per country")
plt.gca().invert_yaxis()
plt.show()
```

## Step 14 — Reading category JSON files

In this step, all JSON files containing video category information were loaded.  
Each JSON file corresponds to a specific country (e.g. `US_category_id.json`),  
and includes a mapping from **`category_id`** to a human-readable **`category_title`**.

For each country:

- the JSON file was opened and parsed,
- the `items` list was extracted,
- a small dataframe with `category_id` and `category_title` was created,
- and all category identifiers were normalized to string type.

The resulting data was stored in a dictionary, mapping the country code to its category dataframe.  
This category information is then used in Step 15 to check whether a video’s category is assignable or not.

```python
# Step 14 - Read all JSON files with the video categories

# Because the notebook and trendingYT folder are in the same directory:
data_path = "../trendingYT"

category_files = glob.glob(os.path.join(data_path, "*_category_id.json"))

print("Found category JSON files:")
for f in category_files:
    print(" -", f)

# Dictionary to store category info per country
categories_by_country = {}

for file in category_files:
    # Extract country code from filename (first two letters, e.g., US, GB, DE)
    country_code = os.path.basename(file)[:2].upper()
    
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    items = data.get("items", [])

    rows = []
    for item in items:
        cat_id = item.get("id")
        title = item.get("snippet", {}).get("title")
        rows.append({"category_id": cat_id, "category_title": title})
    
    categories_df = pd.DataFrame(rows)
    categories_df["category_id"] = categories_df["category_id"].astype(str)
    
    categories_by_country[country_code] = categories_df

# Quick check for one country (e.g., US)
categories_by_country.get("US", pd.DataFrame()).head()
```

## Step 15 — Count videos with non-assignable categories per country

Using the category information loaded in Step 14, this step determines, for each country,  
how many videos have a **non-assignable** category.

A category is considered *assignable* if the video's `category_id` appears in the list of valid category IDs  
for that country (as defined in the corresponding JSON file).  
Otherwise, the video is marked as having a non-assignable category.

A new boolean column, `category_assignable`, was added to the main dataframe.  
Then, for each country, the number of videos with `category_assignable = False` was counted.

This highlights potential inconsistencies or missing category mappings in the dataset across different countries.

```python
# Step 15 - For each country, count videos with a non-assignable category

# First, make sure category_id in all_videos is a string (to match JSON)
all_videos["category_id"] = all_videos["category_id"].astype(str)

# Build a mapping: country -> set of valid category_ids
valid_category_ids = {}
for country_code, df_cat in categories_by_country.items():
    valid_category_ids[country_code] = set(df_cat["category_id"].unique())

# Function to check if a row has an assignable category
def is_category_assignable(row):
    country = row["country"]
    cat_id = row["category_id"]
    
    valid_ids_for_country = valid_category_ids.get(country, set())
    return cat_id in valid_ids_for_country

# Create a boolean column: True if category is assignable, False otherwise
all_videos["category_assignable"] = all_videos.apply(is_category_assignable, axis=1)

# Count non-assignable videos per country
unassignable_counts = (
    all_videos[~all_videos["category_assignable"]]
    .groupby("country")["video_id"]
    .count()
    .reset_index(name="non_assignable_videos")
    .sort_values("non_assignable_videos", ascending=False)
)

unassignable_counts
```

```python
assignable_counts = (
    all_videos[all_videos["category_assignable"]]
    .groupby("country")["video_id"]
    .count()
    .reset_index(name="assignable_videos")
)
```

```python
plt.figure(figsize=(8,5))
plt.barh(unassignable_counts["country"], unassignable_counts["non_assignable_videos"])
plt.xlabel("Number of videos with non-assignable category")
plt.ylabel("Country")
plt.title("Non-assignable categories per country")
plt.gca().invert_yaxis()
plt.show()
```

### Interpretation

Some countries have a noticeably higher number of videos with non-assignable categories.  
This may be due to missing or outdated category definitions in the corresponding JSON files,  
or inconsistencies between the CSV and JSON sources.
